```python
import pandas as pd

df = pd.read_csv("b1abb33c4f.csv")
df.head()
```

####  1) Na base do SINASC do DataSUS, faça um boxplot da idade da mãe.

```python
import matplotlib.pyplot as plt
import pandas as pd


plt.figure(figsize=(8, 6))  
plt.boxplot(df['IDADEMAE'], vert=False)  
plt.title('Idade da mãe')  
plt.xlabel('Idade')  
plt.grid(True)
```

####  2) Faça um boxplot do peso do bebê

```python
plt.figure(figsize=(8, 6))  
plt.boxplot(df['PESO'])  
plt.title('Peso ao nascer')  
plt.xlabel('Peso')  
plt.grid(True)
```

####  3) Faça um histograma da idade da mãe

```python
plt.figure(figsize=(8, 6))  
plt.hist(df['IDADEMAE'], bins=30, color='skyblue', edgecolor='black')
plt.title('Idade da mãe')
plt.xlabel('Idade')
plt.ylabel('Frequência') 
plt.grid(True)
plt.show()
```
