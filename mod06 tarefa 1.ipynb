####  2) Monte um gráficos de barras com as variáveis APGAR1, APGAR5, SEXO, GESTACAO, ESTCIVMAE.

```python
import pandas as pd
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
df['APGAR1'].value_counts().plot(kind='bar', color='skyblue')
plt.title('APGAR1')
plt.xlabel('Apgar')
plt.ylabel('Frequencia')
plt.show()

plt.figure(figsize=(10, 6))
df['APGAR5'].value_counts().plot(kind='bar', color='salmon')
plt.title('APGAR5')
plt.xlabel('Apgar')
plt.ylabel('Frequencia')
plt.show()

plt.figure(figsize=(8, 5))
df['SEXO'].value_counts().plot(kind='bar', color='lightgreen')
plt.title('Sexo')
plt.xlabel('Gênero')
plt.ylabel(' ')
plt.show()

plt.figure(figsize=(10, 6))
df['GESTACAO'].value_counts().plot(kind='bar', color='lightcoral')
plt.title('GESTACAO')
plt.xlabel('Gestacao')
plt.ylabel('Frequencia')
plt.show()

plt.figure(figsize=(10, 6))
df['ESTCIVMAE'].value_counts().plot(kind='bar', color='skyblue')
plt.title('ESTCIVMAE')
plt.xlabel('Estado civil da mãe')
plt.ylabel(' ')
plt.show()
```

####  4) Monte um painel com 4 gráficos mostrando as frequencias de SEXO, GESTACAO, ESTCIVMAE e CONSULTAS

```python
import pandas as pd
import matplotlib.pyplot as plt

# Supondo que você tenha um DataFrame chamado 'df' com as colunas SEXO, GESTACAO, ESTCIVMAE e CONSULTAS

# Configurações para os subplots
fig, axs = plt.subplots(2, 2, figsize=(12, 10))

# Frequência de SEXO
df['SEXO'].value_counts().plot(kind='bar', ax=axs[0, 0], color='skyblue')
axs[0, 0].set_title('SEXO')

# Frequência de GESTACAO
df['GESTACAO'].value_counts().plot(kind='bar', ax=axs[0, 1], color='orange')
axs[0, 1].set_title('GESTACAO')

# Frequência de ESTCIVMAE
df['ESTCIVMAE'].value_counts().plot(kind='bar', ax=axs[1, 0], color='green')
axs[1, 0].set_title('Estado Civíl')

# Frequência de CONSULTAS
df['CONSULTAS'].value_counts().plot(kind='bar', ax=axs[1, 1], color='red')
axs[1, 1].set_title('Frequência de consultas')

# Ajustes de layout
plt.tight_layout()

# Exibir os gráficos
plt.show()
```
